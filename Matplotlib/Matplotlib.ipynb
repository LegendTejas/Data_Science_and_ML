### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='Thick Marked Line',color='red', linewidth= 4, marker='.', markersize=10, markeredgecolor='black')
plt.legend()  # Shows the legend(i.e. for label='Thick Marked Line')

plt.title("Linear Graph", fontdict = {'fontname': 'Monospace', 'fontsize': 15})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])


plt.show()
```

```python
#line 1
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y, 'r^--', markeredgecolor='black', label = '2x')

# Line 2
x2 = np.arange(0, 4.5, 0.5)
y2 = x2**2

# Plot part of the graph for line 2
plt.plot(x2[:8], y2[:8], 'b', label = 'X^2')

# add a title (specify font parameters with fontdict)
plt.title("Linear Graph", fontdict = {'fontname': 'Monospace', 'fontsize': 15})

# X and Y labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# X axis tickmark (scale of your graph)
plt.xticks([0,1,2,3,4,5])

plt.legend() # Show the legend(i.e. for label='...')
plt.show() #Show plot
```

**Resizing Graph**

Syntax: `plt.figure(figsize=(width, height))`

We also use dpi (Dots Per Inch) in plt.figure() to controls the resolution of your figure (i.e., how sharp and detailed the plot looks when displayed or saved.)

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize your graph
plt.figure(figsize=(7,4), dpi=300) # dpi=300 means High-res image
plt.plot(x,y, 'r^--', markeredgecolor='black', label = '2x')

# Line Number Two
x2 = np.arange(0, 4.5, 0.5)
y2 = x2**2
plt.plot(x2, y2, 'b', label = 'X^2')

plt.title("Linear Graph", fontdict = {'fontname': 'Monospace', 'fontsize': 15})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])

plt.legend()
plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]

plt.figure(figsize = (7,4))
bars = plt.bar(labels, values)

# Apply a hatch pattern (diagonal lines '/') to the first bar (bar at index 0) and similarly for other bar
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

patterns = ['/','O','*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

plt.show()
```

## Gas Data

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

###Line Graph

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time (in USD)')

plt.plot(gas.Year, gas.USA,  'b.-', label='USA')
plt.plot(gas.Year,gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-')
## You can also write like this:
# plt.plot(gas['Year'], gas['USA'], label='USA')
# plt.plot(gas['Year'],gas['Canada'], label='Canada')
# plt.plot(gas['Year'], gas['South Korea'], label='South Korea')

plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time (in USD)')

countries = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
  if country in countries:
    plt.plot(gas.Year, gas[country], marker='.', label= country)

plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.title('Gas Prices Over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize':18})

#To directly get all the country gas value in a single graph at once
for country in gas:
  if country != 'Year':
    plt.plot(gas.Year, gas[country], marker='.', label= country)

plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

## FIFA Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
fifa.drop(['Unnamed: 0','Photo','Flag','ID','Club Logo'], axis=1, inplace=True)
fifa
```

###Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='lightgreen')

plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Overall Rating")
plt.title("Distribution of Player Overall Rating in FIFA 2018")
plt.show()
```

###Pie Charts

```python
fifa['Preferred Foot']
```

```python
left = fifa[fifa['Preferred Foot'] == 'Left'].shape[0]
right = fifa[fifa['Preferred Foot'] == 'Right'].shape[0]

print("Left Footer: ",left)
print("Right Footer: ",right)
```

```python
#Left Foot vs Right Foot
labels = ['left','right']
colors = ['skyblue','yellow']
plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')
plt.show()
```

**Explanation of each argument in this line:**

`plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f %%')`

1. `labels=labels:`

* This assigns labels to each pie slice.
* For example: `labels = ['Left Foot', 'Right Foot']`

2. `colors=colors:`
* This sets the colors of the slices.
* Example: `colors = ['skyblue', 'lightgreen']`

3. `autopct='%.2f %%':`

* This displays the percentage value on each slice.

* `%.2f` formats the percentage to 2 decimal places.

`%%` escapes the percent sign (%) so it appears correctly on t

```python
fifa.Weight
```

```python
#removing lbs from all weight
fifa.Weight = [x.strip('lbs') if type(x) == str else x for x in fifa.Weight]
```

```python
print(fifa.Weight)
type(fifa.Weight[0])
```

```python
#NOTE: These weight are in pounds
#convert str value into int
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot') #to use different color scheme for pie chart

light = fifa.loc[fifa.Weight < 125].shape[0]
light_medium =   fifa.loc[(fifa.Weight >= 126) & (fifa.Weight < 150)].shape[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].shape[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].shape[0]
heavy = fifa[(fifa.Weight >= 200)].shape[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150','150-175', '175-200', 'Over 200']
explode = (0.4, .1, 0, 0, 0.4)

plt.title('Weight Distribution of FIFA Players(in lbs)')

plt.pie(weights, labels = labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

```python
#let's style these boxes
plt.style.use('default')

plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
real_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
man_utd = fifa.loc[fifa.Club == 'Manchester United']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Manchester United']

boxes = plt.boxplot([barcelona, real_madrid, man_utd], tick_labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
  #set edge color
  box.set(color= 'purple', linewidth=2)

  #change fill color
  box.set(facecolor='#e0e0e0')
plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
